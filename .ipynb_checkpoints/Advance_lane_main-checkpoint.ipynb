## Advanced Lane Finding Project

The goals / steps of this project are the following:

* Compute the camera calibration matrix and distortion coefficients given a set of chessboard images.
* Apply a distortion correction to raw images.
* Use color transforms, gradients, etc., to create a thresholded binary image.
* Apply a perspective transform to rectify binary image ("birds-eye view").
* Detect lane pixels and fit to find the lane boundary.
* Determine the curvature of the lane and vehicle position with respect to center.
* Warp the detected lane boundaries back onto the original image.
* Output visual display of the lane boundaries and numerical estimation of lane curvature and vehicle position.

## Camera Calibration


````python
objp = np.zeros((6*9,3), np.float32)
objp[:,:2] = np.mgrid[0:9,0:6].T.reshape(-1,2)

# Arrays to store object points and image points from all the images.
objpoints = [] # 3d points in real world space
imgpoints = [] # 2d points in image plane.

# Make a list of calibration images
images = glob.glob('camera_cal/calibration*.jpg')

for path in images:
    img=cv2.imread(path)
    gray=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
    
    ret, corners=cv2.findChessboardCorners(gray,(9,6),None)
    
    
    if ret==True:
        objpoints.append(objp)
        imgpoints.append(corners)

#finding the camera matrix
ret, cameraMatrix, distortionCoeffs, rvecs, tvecs = cv2.calibrateCamera(objpoints, 
                                                                        imgpoints, 
                                                                        img.shape[:2],None,None)

        
````

## Image Thresholding pipeline

````python
def abs_sobel_thresh(img, orient='x', thresh_min=0, thresh_max=255):
    # Convert to grayscale
    gray = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
    # Apply x or y gradient with the OpenCV Sobel() function
    # and take the absolute value
    if orient == 'x':
        abs_sobel = np.absolute(cv2.Sobel(gray, cv2.CV_64F, 1, 0))
    if orient == 'y':
        abs_sobel = np.absolute(cv2.Sobel(gray, cv2.CV_64F, 0, 1))
    # Rescale back to 8 bit integer
    scaled_sobel = np.uint8(255*abs_sobel/np.max(abs_sobel))
    # Create a copy and apply the threshold
    binary_output = np.zeros_like(scaled_sobel)
    # Here I'm using inclusive (>=, <=) thresholds, but exclusive is ok too
    binary_output[(scaled_sobel >= thresh_min) & (scaled_sobel <= thresh_max)] = 1

    # Return the result
    return binary_output

def mag_thresh(img, sobel_kernel=3, mag_thresh=(0, 255)): #30,100
    # Convert to grayscale
    gray = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
    # Take both Sobel x and y gradients
    sobelx = cv2.Sobel(gray, cv2.CV_64F, 1, 0, ksize=sobel_kernel)
    sobely = cv2.Sobel(gray, cv2.CV_64F, 0, 1, ksize=sobel_kernel)
    # Calculate the gradient magnitude
    gradmag = np.sqrt(sobelx**2 + sobely**2)
    # Rescale to 8 bit
    scale_factor = np.max(gradmag)/255 
    gradmag = (gradmag/scale_factor).astype(np.uint8) 
    # Create a binary image of ones where threshold is met, zeros otherwise
    binary_output = np.zeros_like(gradmag)
    binary_output[(gradmag >= mag_thresh[0]) & (gradmag <= mag_thresh[1])] = 1

    # Return the binary image
    return binary_output


def dir_threshold(img, sobel_kernel=3, thresh=(0, np.pi/2)):
    # Grayscale
    gray = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
    # Calculate the x and y gradients
    sobelx = cv2.Sobel(gray, cv2.CV_64F, 1, 0, ksize=sobel_kernel)
    sobely = cv2.Sobel(gray, cv2.CV_64F, 0, 1, ksize=sobel_kernel)
    # Take the absolute value of the gradient direction, 
    # apply a threshold, and create a binary image result
    absgraddir = np.arctan2(np.absolute(sobely), np.absolute(sobelx))
    binary_output =  np.zeros_like(absgraddir)
    binary_output[(absgraddir >= thresh[0]) & (absgraddir <= thresh[1])] = 1

    # Return the binary image
    return binary_output
    
# Define a function that thresholds the S-channel of HLS 90 255
def hls_select(img, s_thresh=(0, 255),l_thresh=(0,255)):
    hls = cv2.cvtColor(img, cv2.COLOR_RGB2HLS)
    s_channel = hls[:,:,2]
    l_channel = hls[:,:,1]
    binary_output1 = np.zeros_like(s_channel)
    binary_output2 = np.zeros_like(s_channel)
    binary_output1[(s_channel > s_thresh[0]) & (s_channel <= s_thresh[1])] = 1
    binary_output2[(l_channel > l_thresh[0]) & (l_channel <= l_thresh[1])] = 1
    return binary_output1,binary_output2  

def rgb_select(img,color_threshold=255):
    R=img[:,:,0]
    G=img[:,:,1]
    rgb_binary=np.zeros_like(R)
    rgb_binary[(R > color_threshold) & (G > color_threshold)]=1
    return rgb_binary




def get_roi_image(img,combined):
    h,w,c=img.shape
    #print(img.shape)
    #genrating black image
    black=np.zeros((h,w)).astype('int8')
    #print(combined.shape)
    # defining some poin to get region of intresed area
    points=[(0,h-1),
         (w/2, int(0.5*h)),
         (w-1, h-1),]
    ver=np.array(points,dtype=np.int32)
    cv2.fillPoly(black, [ver], 255)
    #print(black.shape)
    thresholded = cv2.bitwise_and(combined.astype('int8'), black)
    return thresholded


````

there are lot's of thresholding technique above i have tries but combination of some is giving better result and properly selecting while and yellow line. I have implemented a pipeline for that which is given below

````python
def image_process(image):
    image=cv2.GaussianBlur(image,ksize=(3,3),sigmaX=0)
    gradx = abs_sobel_thresh(image, orient='x', thresh_min=20, thresh_max=200)
    grady = abs_sobel_thresh(image, orient='y', thresh_min=40,thresh_max=100)
    dir_binary = dir_threshold(image, thresh=(np.pi/6,np.pi/2))
    rgb_binary=rgb_select(image,color_threshold=140)
    s_binary,l_binary=hls_select(image,s_thresh=(100, 255),l_thresh=(120, 255))
    binary_img=np.zeros_like(gradx)
    binary_img[((gradx==1) | (grady==1))&(dir_binary==1)|(s_binary==1) & ((l_binary==1) & (rgb_binary==1))]=1
    
    return get_roi_image(image,binary_img) 
````

````python
def get_warped_image(thres_binary,M):
    warped = cv2.warpPerspective(thres_binary.astype('float32'), M, image_size,flags=cv2.INTER_LINEAR )
    return warped

def get_base_points(warped):
    histogram=np.sum(warped[warped.shape[0]//2:,:],axis=0)
    center=len(histogram)//2
    leftx_base=np.argmax(histogram[:center])

    rightx_base=np.argmax(histogram[center:])+center
    
    return leftx_base,rightx_base
    
````

## Sliding Window Search

````python
def window_search(warped,leftx_base,rightx_base):
    out_img = np.dstack((warped, warped, warped))*255

    non_zeros = warped.nonzero()
    non_zeros_y = non_zeros[0]
    non_zeros_x = non_zeros[1]

    num_windows = 10
    num_rows = warped.shape[0]
    window_height = np.int(num_rows/num_windows)
    window_half_width = 100

    min_pixels = 100

    left_coordinates = []
    right_coordinates = []

    
    #drawing and seaching line in this loop
    for window in range(num_windows):

        
        #making bourdries for windows
        y_max = num_rows - window*window_height
        y_min = num_rows - (window+1)* window_height

        
        
        
        left_x_min = leftx_base - window_half_width
        left_x_max = leftx_base + window_half_width
        
    
        #drawing window on the warped image
        cv2.rectangle(out_img, (left_x_min, y_min), (left_x_max, y_max), [0,0,255],2)

        
        #searching for nonzero or white pixel within the boundry if exist then take those cordinates
        good_left_window_coordinates = ((non_zeros_x >= left_x_min) & (non_zeros_x <= left_x_max) & (non_zeros_y >= y_min) & (non_zeros_y <= y_max)).nonzero()[0]
        left_coordinates.append(good_left_window_coordinates)

        
        #if number of pixel is more than the min_pixel threshold then it will considered as part of line
        if len(good_left_window_coordinates) > min_pixels:
            leftx_base = np.int(np.mean(non_zeros_x[good_left_window_coordinates]))



        #making boundries for right side lane and same procedure which we are following just above code
        right_x_min = rightx_base - window_half_width
        right_x_max = rightx_base + window_half_width

        cv2.rectangle(out_img, (right_x_min, y_min), (right_x_max, y_max), [0,0,255],2)

        good_right_window_coordinates = ((non_zeros_x >= right_x_min) & (non_zeros_x <= right_x_max) & (non_zeros_y >= y_min) & (non_zeros_y <= y_max)).nonzero()[0]
        right_coordinates.append(good_right_window_coordinates)

        
        #updating the next window position
        if len(good_right_window_coordinates) > min_pixels:
            rightx_base = np.int(np.mean(non_zeros_x[good_right_window_coordinates]))



    #making numpy array from the list     
    left_coordinates = np.concatenate(left_coordinates)
    right_coordinates = np.concatenate(right_coordinates)


    #selecting the nonzero pixel which has passed the min_pixel threshold condition
    #now we are giving color to left line red
    out_img[non_zeros_y[left_coordinates], non_zeros_x[left_coordinates]] = [255,0,0]

    #for the left line color is blue
    out_img[non_zeros_y[right_coordinates], non_zeros_x[right_coordinates]] = [0,0,255]


    #selecting the cordinates of nonzero pixel
    left_x = non_zeros_x[left_coordinates]
    left_y = non_zeros_y[left_coordinates]


    #with the help of left_x and left_y pixel we finding the second order polynomial Ax2+Bx+C
    polyfit_left = np.polyfit(left_y, left_x, 2)

    right_x = non_zeros_x[right_coordinates]
    right_y = non_zeros_y[right_coordinates]

    polyfit_right = np.polyfit(right_y, right_x, 2)
    
    #getting y points
    y_points = np.linspace(0, num_rows-1, num_rows)
    
    return polyfit_left, polyfit_right, y_points, out_img

````

## Searching around the previously detected line in last frame

````python
def normal_search(warped,polyfit_left,polyfit_right):
    margin=50 #we are taking the margin at both side of predicted line

    #making three layer image from the warped single layer image
    out_image=np.dstack((warped,warped, warped))*255

    #selecting those pixel cordintes which has white color or non zero value
    non_zeros = warped.nonzero()
    non_zeros_y = non_zeros[0]
    non_zeros_x = non_zeros[1]


    #alreay we have previous frame polynomial perameters, we gona predict the next polynomial or line aroound
    #it without the applying sliding window method
    predicted_left_line_x=polyfit_left[0]*non_zeros_y**2+polyfit_left[1]*non_zeros_y+polyfit_left[2]

    #adding the margin both side of this predicted line and getting y codrinates of it or you can take x as well
    left_cordinates_y= ((non_zeros_x >= predicted_left_line_x - margin) & (non_zeros_x <= predicted_left_line_x + margin)).nonzero()[0]


    #same for the right line
    predicted_right_line_x=polyfit_right[0]*non_zeros_y**2+polyfit_right[1]*non_zeros_y+polyfit_right[2]
    right_cordinates_y= ((non_zeros_x >= predicted_right_line_x - margin) & (non_zeros_x <= predicted_right_line_x + margin)).nonzero()[0]


    #giving color to the line
    out_image[non_zeros_y[left_cordinates_y], non_zeros_x[left_cordinates_y]] = [255,0,0]
    out_image[non_zeros_y[right_cordinates_y], non_zeros_x[right_cordinates_y]] = [0,0,255]

    #finding white pixel for new polynomial around the previous polynomial
    left_x_cord = non_zeros_x[left_cordinates_y]
    left_y_cord= non_zeros_y[left_cordinates_y]
    #getting new polnomial perameter
    polyfit_left = np.polyfit(left_y_cord, left_x_cord, 2)

    #for right line
    right_x_cord = non_zeros_x[right_cordinates_y]
    right_y_cord= non_zeros_y[right_cordinates_y]
    polyfit_right = np.polyfit(right_y_cord, right_x_cord, 2)
    
    return polyfit_left, polyfit_right, out_image

````

````python
def get_curve_radius(l_x_points,r_x_points,y_points):
    #we have to calibarate the pixel according to camera range
    #as provided width=3.7m
    #height=30m
    ym_per_pix = 30/720 # meters per pixel in y dimension
    xm_per_pix = 3.7/700
    
    
    y_eval = np.max(y_points)
    
    left_fit_cr=np.polyfit(y_points*ym_per_pix, l_x_points*xm_per_pix, 2)
    right_fit_cr=np.polyfit(y_points*ym_per_pix, r_x_points*xm_per_pix, 2)
    
    # Calculation of R_curve (radius of curvature)
    left_curverad = (
        (1 + (2*left_fit_cr[0]*y_eval*ym_per_pix + left_fit_cr[1])**2)**1.5) / np.absolute(2*left_fit_cr[0])
    
    right_curverad = (
        (1 + (2*right_fit_cr[0]*y_eval*ym_per_pix + right_fit_cr[1])**2)**1.5) / np.absolute(2*right_fit_cr[0])
    
    return (left_curverad+right_curverad)/2


def get_center_offset(left_x_predictions,right_x_predictions):
    xm_per_pix = 3.7/700
    center_boundary=(left_x_predictions[-1]+right_x_predictions[-1])/2
    image_center=image.shape[1]/2
    return abs(image_center-center_boundary)*xm_per_pix




avg_radius=get_curve_radius(predicted_left_line_x,predicted_right_line_x,y_points)

#print("the left radius %.2f m"%l_radius)
print("the curveture radius %.2f m"%avg_radius)
c=get_center_offset(left_x_predictions,right_x_predictions)
print("center offset %.2f "%c)
````

````python
def draw_road(predicted_left,predicted_right,M_inverse,original_image,warped_image):
    
    #y_points = np.linspace(0, num_rows-1, num_rows)
    
    out_img=np.zeros_like(original_image)
    #print(predicted_left.shape)
    #print(y_points.shape)
    
    left_points=np.array(np.vstack((predicted_left,y_points)).T)
    
    right_points=np.array(np.flipud(np.vstack((predicted_right,y_points)).T))
    
    final_points=np.vstack((left_points,right_points))
    
    cv2.fillPoly(out_img,final_points.reshape(1,-1,2).astype('int'),(0,255,0))
    
    out= cv2.warpPerspective(out_img, M_inv, image_size,flags=cv2.INTER_LINEAR )
    
   
    
    final_image=cv2.addWeighted(original_image,1,out.astype('uint8'),0.5,0)
    
    curvature="Curveture radius : %.2fm" %get_curve_radius(predicted_left,predicted_right,y_points)
    offset="Center offSet : %.2fm" %get_center_offset(predicted_left,predicted_right)
    
    cv2.putText(final_image, str(curvature) , (100, 90), cv2.FONT_HERSHEY_SIMPLEX, 1.5, (255,255,255), thickness=2)
    cv2.putText(final_image, str(offset), (100, 150), cv2.FONT_HERSHEY_SIMPLEX, 1.5, (255,255,255), thickness=2)
    
    return final_image
    
````

## Final Pipeline 

````python
left_predictions=[]
right_predictions=[]
y_points=[]
polyfit_left=[]
polyfit_right=[]

def pipeline(image):
    
    global left_predictions
    global right_predictions
    global y_points
    global polyfit_left
    global polyfit_right
    
    image=cv2.undistort(image,cameraMatrix,distortionCoeffs,None,cameraMatrix)
    
    thres_binary=image_process(image)
    
    warped_image=get_warped_image(thres_binary,M)
    
    left_base,right_base=get_base_points(warped_image)
    
    

    #if we find some bad frame diffrent frame the we will apply winow search method
    if (len(left_predictions) is 0) or (len(right_predictions) is 0):
        print("runining")
    
        polyfit_left,polyfit_right,y_points,image_d =window_search(warped_image, left_base, right_base)
        

        left_predictions = polyfit_left[0]*y_points**2 + polyfit_left[1]*y_points + polyfit_left[2]

        right_predictions = polyfit_right[0]*y_points**2 + polyfit_right[1]*y_points + polyfit_right[2]
        
    #print(left_predictions)  
    
    #other wise we will do normal line search
    else :
        polyfit_left,polyfit_right,image_d =normal_search(warped_image,polyfit_left,polyfit_right)

        #print("right",polyfit_right)

        left_predictions = polyfit_left[0]*y_points**2 + polyfit_left[1]*y_points + polyfit_left[2]

        right_predictions = polyfit_right[0]*y_points**2 + polyfit_right[1]*y_points + polyfit_right[2]
        
    
    
    img=draw_road(left_predictions,right_predictions,M_inv,image,warped)
    
    return img
````
